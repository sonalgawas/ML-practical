```python
import pandas as pd

data = {
    "name": ["tejas", "suyash", "yash", "nikhil", "bhavesh"],
    "age": [20, 22, 19, 24, 28],
    "score": [87, 99, 93, 80, 85]  # Added a 5th score
}

df = pd.DataFrame(data)
```

```python
print("Dataframe:\n",df)
```

```python
name_column=df["name"]
print("n\Retreived column-name:\n",name_column)
```

```python
summary = df.describe()
print("\nSummary statistics of DataFrame:\n", summary)
```

```python
mean_values=df.select_dtypes(include=[np.number]).mean()
std_values=df.select_dtypes(include=[np.number]).std()
```

```python
print("\nMean values:\n",mean_values)
print("\nStandard deviation values:\n",std_values)
```
