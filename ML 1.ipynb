```python
import pandas as pd
import numpy as np
```

```python
def calculate_mean(numbers):
  return sum(numbers) / len(numbers)
```

```python
def calculate_mean(numbers):
  numbers.sort()
  n = len(numbers)
  middle = n // 2
  return (numbers[middle - 1] + numbers[middle]) / 2 if n % 2 == 0 else numbers[middle]
```

```python
def calculate_standard_devation(numbers):
  mean = calculate_mean(numbers)
  variance = sum((x - mean) ** 2 for x in numbers) / len(numbers)
  return variance ** 0.5
```

```python
def calculate_mode(numbers):
  from collections import Counter
  frequency = Counter(numbers)
  max_count = max(frequency.values())
  return [k for k, values in frequency.items() if values == max_count]
```

```python
# Input and Execution
numbers = list(map(int, input("Enter numbers seprate by spaces:").split()))
```

```python
print(f"Mean: {calculate_mean(numbers)}")
print(f"Median: {calculate_mean(numbers)}")
print(f"Standard Deviation: {calculate_standard_devation(numbers)}")
print(f"Mode:{calculate_mode(numbers)}")
```
